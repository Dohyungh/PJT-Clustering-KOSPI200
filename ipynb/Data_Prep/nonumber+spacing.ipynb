```python
import pandas as pd
import re
```

```python
df = pd.read_pickle('./dart_crawling_result.fkl')
```

```python
df1 = df[['name', 'code', 'x']]
df1
```

```python
spacing_word = ['삼성', '현대', '한화', '키움', '두산', '롯데', '한국', 'SK', 'GS', 'LG', 'CJ', 'LX']
incorrect_word = ['플래쉬', '에스케이', '램', '케이씨씨', '수퍼', 'A/S', 'R&D', 'C&C', 'F&F']
correct_word = ['플래시', 'SK', 'RAM', 'KCC', '슈퍼', 'AS', 'RD', 'CC', '에프앤에프']
stop_word = ['동사', '및', '의', '등', '부문', '주']
```

```python
y=[]
for i in range(200):
    sen = df.iloc[i]['x']
    sen = re.sub('\(|\)', ' ', sen)
    #sen = re.sub('/', '', sen)
    #sen = re.sub('&', '', sen)
    sen1 = sen.split()
    sen2 = ''
    for j in range(len(sen1)):
        temp = (re.sub('\d.*', '', sen1[j]))
        for k in range(len(incorrect_word)):
            temp = re.sub(incorrect_word[k], correct_word[k], temp)
        for k in spacing_word:
            temp = re.sub(k + '.*', k + ' ' + temp[2:], temp)
        sen2 += temp + ' '
    y.append(sen2.upper())
df['x2']= y
df
```
