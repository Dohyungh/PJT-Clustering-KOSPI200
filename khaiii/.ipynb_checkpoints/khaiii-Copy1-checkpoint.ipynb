```python
import pandas as pd
import os
import pickle
import sys
```

```python
from khaiii import KhaiiiApi
api = KhaiiiApi()
```

```python
import pickle5 as pkl
with open('./parsing.fkl', "rb") as fh:
  df1 = pkl.load(fh)
```

```python
df1
```

```python
li = []
for i in range(0,200):
    morphs = []
    for word in api.analyze(df1.iloc[i]['x2']):
        for morph in word.morphs:
            if morph.tag in ['NNG', 'NNP', 'SL']:
                morphs.append(morph.lex)
    li.append(morphs)
df1["morph"]= li

string_list = []
for i in range(0,200):
    sum_string = ""
    for j in range(0,len(df1.iloc[i]["morph"])):
        sum_string+= (df1.iloc[i]["morph"][j]+ " ")
    string_list.append(sum_string)
    
df1["morph_sum"] = string_list
df1
```
